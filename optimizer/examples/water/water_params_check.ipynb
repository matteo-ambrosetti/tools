```python
import matplotlib.pyplot as plt
import numpy as np
```

# Plot both the computed and reference interaction energy for several scans

```python
folder_list = ["water_plane_H", "water_plane_O", "water_scan_H-O-H_bisec", "water_scan_H-O-H_bisec_ext", "water_scan_O-H"]
opt_folder  = "out_w1.0_int0.0-1.0"
for folder in folder_list:
    fig, ax = plt.subplots(figsize=(15,10))
    fig.suptitle(opt_folder, fontsize=35)
    ax.set_title(folder,fontsize= 30) # title of plot
    for i in range(1,11):
        data = np.loadtxt("{}/graph_run{}_top1_{}.csv".format(opt_folder,i,folder))
        ax.plot(data[:,0], data[:,1], label='run number {}'.format(i), linewidth=5)
    ax.plot(data[:,0], data[:,2], label='reference'.format(i), linewidth=8, linestyle='dashed')
    plt.xlabel('distance [$\AA$]', fontsize=20)
    plt.xticks(fontsize=20)
    plt.ylabel('Interaction Energy [kcal/mol]', fontsize=20)
    plt.yticks(fontsize=20)
    ax.grid()
    ax.legend(fontsize='xx-large')
    plt.show()
```
